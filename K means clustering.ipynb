```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
X = College_data.drop('Private',axis=1)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
sns.set_style('darkgrid')
sns.scatterplot(kmeans.cluster_centers_)
```

```python
kmeans.labels_
```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=kmeans.labels_, cmap='rainbow', alpha=0.6, edgecolors='w')
```
